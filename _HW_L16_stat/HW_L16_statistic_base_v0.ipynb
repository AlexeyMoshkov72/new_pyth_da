### Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
Q1_rt = data_short.rectal_temperature.quantile(0.25)
Q3_rt = data_short.rectal_temperature.quantile(0.75)

iqr_rt = Q3_rt - Q1_rt

lower_bound_rt = Q1_rt - (1.5 * iqr_rt) 
upper_bound_rt = Q3_rt + (1.5 * iqr_rt)

remove_outliers_rt = data_short[data_short.rectal_temperature.between(lower_bound_rt, upper_bound_rt, inclusive=True)].sort_values('rectal_temperature')
```

```python
pd.concat([data_short, remove_outliers_rt]).drop_duplicates(keep=False).dropna(subset=['rectal_temperature'])
```

Все значения выбросов находятся в перделах допустимых значений. При использовании данной информации выявленные выбросы следует сохранить.

```python
Q1_p = data_short.pulse.quantile(0.25)
Q3_p = data_short.pulse.quantile(0.75)

iqr_p = Q3_p - Q1_p

lower_bound_p = Q1_p - (1.5 * iqr_p) 
upper_bound_p = Q3_p + (1.5 * iqr_p)

remove_outliers_p = data_short[data_short.pulse.between(lower_bound_p, upper_bound_p, inclusive=True)].sort_values('pulse')
```

```python
pd.concat([data_short, remove_outliers_p]).drop_duplicates(keep=False).dropna(subset=['pulse'])
```

Нормальные значения пульса лошади в состоянии покоя составляют 36-44 ударов в минуту. Представленные выбросы сильно превышают данные значения. Однако, возможно, что данные измерения проводились у лошадей после физических нагрузок. Поэтому эти данные также не следует исключать при анализе.

```python
Q1_tp = data_short.total_protein.quantile(0.25)
Q3_tp = data_short.total_protein.quantile(0.75)

iqr_tp = Q3_tp - Q1_tp

lower_bound_tp = Q1_tp - (1.5 * iqr_tp) 
upper_bound_tp = Q3_tp + (1.5 * iqr_tp)

remove_outliers_tp = data_short[data_short.total_protein.between(lower_bound_tp, upper_bound_tp, inclusive=True)].sort_values('total_protein')
```

```python
pd.concat([data_short, remove_outliers_tp]).drop_duplicates(keep=False).dropna(subset=['total_protein'])
```

Как было указано ранее, по данному показателю наблюдался большой разброс между значениями 1-ого и 3-его квартилей. В результате выбросов по данному показателю нет.

### Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for col in data_short.columns:
    pct_missing = data_short[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

Удаляем неполностью заполненые строки (это незначительно сокращает объем информации с 300 строк до 282 строк)

```python
data_short_full = data_short.dropna(thresh=7)
data_short_full.info()
```

```python
for col in data_short_full.columns:
    pct_missing = data_short_full[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

В столбцах где пропущено незначительное количество значений, а также содержатся категориальные показатели, пропущенные значения заменяем модой.
Такая замена не окажет существенное влияние на возможные выводы при анализе этих данных.

```python
data_short_full.surgery.fillna(data_short_full.surgery.mode()[0], inplace=True)
data_short_full.outcome.fillna(data_short_full.outcome.mode()[0], inplace=True)

data_short_full
```

Для показателей, где количество пропусков достаточно сущетсвенно, и не определены взаимосвязи между этими показателями, пустые значения заменяем на прямо указание на остутствие информации

```python
data_short_full.pulse.fillna('no_info', inplace=True)
data_short_full.total_protein.fillna('no_info', inplace=True)
data_short_full.abdominal_distension.fillna('no_info', inplace=True)
data_short_full.pain.fillna('no_info', inplace=True)
data_short_full.rectal_temperature.fillna('no_info', inplace=True)
```

Для показателя nasogastric_reflux пропуски заполняем значениями медианы, в зависимости от показателя abdominal_distension.

```python
data_short_full.nasogastric_reflux.fillna(data_short_full.groupby(['abdominal_distension']).nasogastric_reflux.transform('median'), inplace=True)
```

```python
for col in data_short_full.columns:
    pct_missing = data_short_full[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```
