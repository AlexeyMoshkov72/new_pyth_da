### Задание 1**

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

**Домашнее задание к лекции.

```python
import pandas as pd
```

```python
df_rat = pd.read_csv('D:\\Documents_AL\\Netology\\Python\\ml-latest-small\\ratings.csv')
df_rat.head()
```

```python
def rating_class(rating):
    if rating <= 2:
        rat_class = 'low'
    elif 2 < rating <= 4:
        rat_class = 'midlle'
    elif rating > 4:
        rat_class = 'high'
    
    return rat_class
```

```python
df_rat['class'] = df_rat['rating'].apply(rating_class)
df_rat.head()
```

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

Важно! При копировании юпитер ноутбука изменяет кавычки ’ на ` , что может способствовать ошибки в коде. В данной ситуации необходимо их просто заменить.

```python
import os
import pandas as pd

data_mov = pd.read_csv('D:\\Documents_AL\\Netology\\Python\\ml-latest\\movies.csv')

data_mov.head()
```

```python
years = []
for year in range(1950, 2011):
    years.append(str(year))
```

```python
def production_year(title):
    for year in range(1950, 2011):
        if str(year) in title:
            return year
    return 1900
```

```python
# добавляем столбец с годом выпуска фильма
data_mov['year'] = data_mov['title'].apply(production_year)

# убираем лишние столбцы из данных
data_mov_year = data_mov[['movieId','title','year']]

data_mov_year.head()
```

```python
data_rat = pd.read_csv('D:\\Documents_AL\\Netology\\Python\\ml-latest\\ratings.csv')

#убираем лишние столбцы из данных
data_rat = data_rat[['movieId', 'rating']]

#считаем средний рейтинг каждого фильма
data_rat = data_rat.groupby(['movieId']).rating.mean()

data_rat.head()
```

```python
#переводим расчет среднего рейтинга каждого фильма в DataFrame
df_av_rating = pd.DataFrame({'movieId':data_rat.index, 'av_rating':data_rat.values})
df_av_rating.head()
```

```python
#соединяем 2 таблицы. 1 - таблица содержащая год выпуска фильма, 2 - таблица содержащая средний рейтинг каждого фильма
df_united = df_av_rating.merge(data_mov_year, how = 'left')

df_united.head()
```

```python
# расчитыаем средний рейтинг фильмов, выпущенных в каждом году
av_rating_year = df_united.groupby(['year']).av_rating.mean()

av_rating_year
```

```python
#формируем DataFrame, содержащий средний рейтинг фильмов для каждого года
df_res = pd.DataFrame({'year':av_rating_year.index, 'av_rating_year':av_rating_year.values})

# сортируем данные по среднему рейтингу за год по убыванию
df_res.sort_values('av_rating_year', ascending=False).head()
```
